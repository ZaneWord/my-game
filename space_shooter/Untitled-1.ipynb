```python
<!DOCTYPE html>
<html>
<head>
    <title>太空射击游戏</title>
    <meta name="viewport" content="width=device-width, initial-scale=1.0, maximum-scale=1.0, user-scalable=no">
    <style>
        /* 新增重启按钮样式 */
        #restartButton {
            position: fixed;
            top: 60%;
            left: 50%;
            transform: translateX(-50%);
            padding: 12px 24px;
            font-size: 20px;
            background: #4CAF50;
            color: white;
            border: none;
            border-radius: 5px;
            cursor: pointer;
            display: none;
        }
    </style>
</head>
<body>
    <canvas id="gameCanvas"></canvas>
    <button id="restartButton">重新开始</button>
    <script>
        const canvas = document.getElementById('gameCanvas');
        const ctx = canvas.getContext('2d');
        const restartButton = document.getElementById('restartButton');

        // 游戏状态初始化函数
        function initGame() {
            return {
                score: 0,
                isGameOver: false,
                player: {
                    x: canvas.width/2,
                    y: canvas.height - 100,
                    size: 30,
                    speed: 5,
                    isMovingLeft: false,
                    isMovingRight: false
                },
                bullets: [],
                enemies: [],
                explosions: [],
                lastShotTime: 0  // 新增射击时间控制
            };
        }

        let gameState = initGame();

        // 设置画布尺寸
        function resizeCanvas() {
            canvas.width = window.innerWidth;
            canvas.height = window.innerHeight;
        }
        window.addEventListener('resize', resizeCanvas);
        resizeCanvas();

        // 游戏主循环
        function gameLoop() {
            if(gameState.isGameOver) return;
            
            update();
            draw();
            requestAnimationFrame(gameLoop);
        }

        function update() {
            // 限制子弹和敌人数量，避免性能问题
            if(gameState.bullets.length > 50) {
                gameState.bullets = gameState.bullets.slice(-50);
            }
            if(gameState.enemies.length > 20) {
                gameState.enemies = gameState.enemies.slice(-20);
            }
            if(gameState.explosions.length > 10) {
                gameState.explosions = gameState.explosions.slice(-10);
            }

            // 玩家移动
            if(gameState.player.isMovingLeft) {
                gameState.player.x = Math.max(gameState.player.size, gameState.player.x - gameState.player.speed);
            }
            if(gameState.player.isMovingRight) {
                gameState.player.x = Math.min(canvas.width - gameState.player.size, gameState.player.x + gameState.player.speed);
            }

            // 自动射击（每100毫秒发射一次）
            const now = Date.now();
            if(now - gameState.lastShotTime > 100) {
                gameState.bullets.push({
                    x: gameState.player.x,
                    y: gameState.player.y,
                    size: 5,
                    speed: -8
                });
                gameState.lastShotTime = now;
            }

            // 更新子弹
            gameState.bullets = gameState.bullets.filter(bullet => {
                bullet.y += bullet.speed;
                return bullet.y > 0;
            });

            // 动态敌人速度：基础速度1.5，每50分增加0.3
            const baseSpeed = 1.5 + Math.floor(gameState.score / 50) * 0.3;

            // 生成敌人
            if(Math.random() < 0.02 && !gameState.isGameOver) {
                const minDistance = 100; // 最小生成距离
                let newEnemyX;
                let isTooClose;
                
                do {
                    newEnemyX = Math.random() * (canvas.width - 40) + 20;
                    isTooClose = gameState.enemies.some(enemy => 
                        Math.abs(enemy.x - newEnemyX) < minDistance
                    );
                } while(isTooClose);

                gameState.enemies.push({
                    x: newEnemyX,
                    y: 0,
                    size: 20,
                    speed: baseSpeed
                });
            }
    
            // 更新敌人和碰撞检测
            gameState.enemies = gameState.enemies.filter(enemy => {
                enemy.y += enemy.speed;
                
                // 子弹碰撞检测
                let hit = false;
                gameState.bullets = gameState.bullets.filter(bullet => {
                    const dx = bullet.x - enemy.x;
                    const dy = bullet.y - enemy.y;
                    const distance = Math.sqrt(dx*dx + dy*dy);
                    
                    if(distance < bullet.size + enemy.size) {
                        hit = true;
                        gameState.score += 10;
                        gameState.explosions.push({
                            x: enemy.x,
                            y: enemy.y,
                            radius: 0,
                            maxRadius: 30
                        });
                        return false; // 移除子弹
                    }
                    return true;
                });

                if(hit) return false; // 被击中的敌人消失

                // 玩家碰撞检测
                const dx = gameState.player.x - enemy.x;
                const dy = gameState.player.y - enemy.y;
                if(Math.sqrt(dx*dx + dy*dy) < gameState.player.size + enemy.size) {
                    gameOver();
                    return false;
                }

                return enemy.y < canvas.height;
            });

            // 更新爆炸效果
            gameState.explosions = gameState.explosions.filter(explosion => {
                explosion.radius += 2;
                return explosion.radius < explosion.maxRadius;
            });
        }

        function draw() {
            ctx.fillStyle = '#000';
            ctx.fillRect(0, 0, canvas.width, canvas.height);

            // 绘制玩家
            ctx.fillStyle = '#0f0';
            ctx.beginPath();
            ctx.arc(gameState.player.x, gameState.player.y, gameState.player.size, 0, Math.PI * 2);
            ctx.fill();

            // 绘制子弹
            ctx.fillStyle = '#ff0';
            gameState.bullets.forEach(bullet => {
                ctx.beginPath();
                ctx.arc(bullet.x, bullet.y, bullet.size, 0, Math.PI * 2);
                ctx.fill();
            });

            // 绘制敌人
            ctx.fillStyle = '#f00';
            gameState.enemies.forEach(enemy => {
                ctx.beginPath();
                ctx.arc(enemy.x, enemy.y, enemy.size, 0, Math.PI * 2);
                ctx.fill();
            });

            // 绘制爆炸
            gameState.explosions.forEach(explosion => {
                ctx.strokeStyle = `rgba(255, 165, 0, ${1 - explosion.radius/explosion.maxRadius})`;
                ctx.beginPath();
                ctx.arc(explosion.x, explosion.y, explosion.radius, 0, Math.PI * 2);
                ctx.stroke();
            });

            // 绘制分数
            ctx.fillStyle = '#fff';
            ctx.font = '24px Arial';
            ctx.fillText(`得分: ${gameState.score}`, 10, 30);
        }

        function gameOver() {
            gameState.isGameOver = true;
            ctx.fillStyle = 'rgba(0, 0, 0, 0.7)';
            ctx.fillRect(0, 0, canvas.width, canvas.height);
            ctx.fillStyle = '#fff';
            ctx.font = '48px Arial';
            ctx.textAlign = 'center';
            ctx.fillText('游戏结束', canvas.width/2, canvas.height/2);
            ctx.font = '24px Arial';
            ctx.fillText(`最终得分: ${gameState.score}`, canvas.width/2, canvas.height/2 + 40);
            
            // 显示重启按钮
            restartButton.style.display = 'block';
        }

        // 重启游戏
        restartButton.addEventListener('click', () => {
            gameState = initGame();
            restartButton.style.display = 'none';
            gameLoop();
        });

        // 控制事件（移除射击相关控制）
        window.addEventListener('keydown', (e) => {
            if(e.key === 'ArrowLeft') gameState.player.isMovingLeft = true;
            if(e.key === 'ArrowRight') gameState.player.isMovingRight = true;
        });

        window.addEventListener('keyup', (e) => {
            if(e.key === 'ArrowLeft') gameState.player.isMovingLeft = false;
            if(e.key === 'ArrowRight') gameState.player.isMovingRight = false;
        });

        // 触摸控制
        canvas.addEventListener('touchstart', (e) => {
            const touchX = e.touches[0].clientX;
            if(touchX < canvas.width/2) {
                gameState.player.isMovingLeft = touchX < gameState.player.x;
                gameState.player.isMovingRight = !gameState.player.isMovingLeft;
            }
            e.preventDefault();
        });

        canvas.addEventListener('touchend', (e) => {
            gameState.player.isMovingLeft = false;
            gameState.player.isMovingRight = false;
            e.preventDefault();
        });

        // 开始游戏
        gameLoop();
    </script>
</body>
</html>
```
